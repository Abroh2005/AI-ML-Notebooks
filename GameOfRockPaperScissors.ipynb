```python
# Import Random
import random

# Declare Variables
global count_Rock#0
global count_Scissor#1
global count_Paper#2
global user_input
global pred
global computer_Score
global player_Score
# Initialize Variables
count_Rock=count_Scissor=count_Paper=pred=computer_Score=player_Score=user_input=0

# update count()
def update_Count(user_input):
  if user_input==0:
    count_Rock+=1
  elif user_input==1:
    count_Scissor+=1
  elif user_input==2:
    count_Scissor+=1
  else:
    print("Invalid Input")

# prediction()
def prediction():
  if count_Rock>count_Paper and count_Rock>count_Scissor:
    pred=0
  if count_Scissor>count_Paper and count_Scissor>count_Rock:
    pred=1
  if count_Paper>count_Rock and count_Paper>count_Scissor:
    pred=2
  return pred

#update scores()
#0 over 1,1 over 2,2 over 0
def update_Scores(user_input):
  pred=prediction()
  if user_input==0:
    if pred==0:
      print("\nYou played Rock and computer played Rock\n")
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
    elif pred==1:
      print("\nYou played Rock and computer played Scissor\n")
      player_Score+=1
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
    else:
      print("\nYou played Rock and computer played Paper\n")
      computer_Score+=1
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
  elif user_input==1:
    if pred==0:
      print("\nYou played Scissor and computer played Rock\n")
      computer_Score+=1
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
    elif pred==1:
      print("\nYou played Scissor and computer played Scissor\n")
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
    else:
      print("\nYou played Scissor and computer played Paper\n")
      player_Score+=1
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
  elif user_input==2:
    if pred==0:
      print("\nYou played Paper and computer played Rock\n")
      player_Score+=1
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
    elif pred==1:
      print("\nYou played Paper and computer played Scissor\n")
      computer_Score+=1
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
    else:
      print("\nYou played Paper and computer played Paper\n")
      print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")
  else:
    print("Foul")
    print("Computer Score = ",computer_Score,"\tYour Score = ",player_Score,"\n")

#gameplay()
def gameplay():
  valid_inputs=[0,1,2]
  print("Hello Player!!!\nWelcome to the Ultimate Game of ROCK, PAPER, SCISSORS\nNow you will be playing against the commputer hehehe\nTime to Spice it UP!!\n");
  user_input=input("Enter 0 for ROCK, 1 for SCISSOR and 2 for PAPER")
  while user_input not in valid_inputs:
    print("Invalid Input\nPlease Enter a Valid Input")
    user_input=input("Enter 0 for ROCK, 1 for SCISSOR and 2 for PAPER")
  user_input=int(user_input)
  update_Scores(user_input)
  update_Count(user_input)

#main body
while True:
  gameplay()
  if computer_Score==10:
    break
  if player_Score==10:
    break
```
