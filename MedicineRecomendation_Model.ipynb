```python
# Initializing LabelEncoder
le = LabelEncoder()

# Fit and transform the target variable
yencoded = le.fit_transform(y)
```

```python
yencoded
```

```python
# Split the data set
xtrain,xtest,ytrain,ytest=train_test_split(x,yencoded,test_size = 0.2, random_state = 0)
```

# Logistic Regretion

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize Logistic Regretion model
lr_model = LogisticRegression()

# Train the model n training dta set
lr_model.fit(xtrain, ytrain)

# Make prediction on the train data set
y_train_pred_lr = lr_model.predict(xtrain)

# Make prediction on the test data set
y_test_pred_lr = lr_model.predict(xtest)

# Evaluate the model
print('Training Accuracy on Train Dataset:', accuracy_score(ytrain, y_train_pred_lr))
print('Training Accuracy on Test Dataset:', accuracy_score(ytest, y_test_pred_lr))

# Display the confusion matrix and the classification report for test dataset
print('\nConfusion Matrix:', confusion_matrix(ytest, y_test_pred_lr))
print('\nClassification Report:', classification_report(ytest, y_test_pred_lr))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize the Random Forest Classifier Model
rf_model = RandomForestClassifier()

# Train the model on train dataset
rf_model.fit(xtrain, ytrain)

# Make prediction on the train data set
y_train_pred_rf = rf_model.predict(xtrain)
# Mkae prediction on the test data set
y_test_pred_rf = rf_model.predict(xtest)

# Evaluate the model
print('Training Accuracy on the train dataset:', accuracy_score(ytrain, y_train_pred_rf))
print('Training Accuracy on the test dataset:', accuracy_score(ytest, y_test_pred_rf))

# Display the confusion matrix and classification report for test dataset
print('\nConfusion Matrix:', confusion_matrix(ytest, y_test_pred_rf))
print('\nClassification Report:', classification_report(ytest, y_test_pred_rf))
```

# Support Vector Classifier

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize the Support Vector Classifier Model
svc_model = SVC()

# Train the model on the training dataset
svc_model.fit(xtrain, ytrain)

# Make prediction on the train data set
y_train_pred_svc = svc_model.predict(xtrain)
# Make prediction on the test data set
y_test_pred_svc = svc_model.predict(xtest)

# Evaluate the model
print('Training Accuracy on the train dataset:', accuracy_score(ytrain, y_train_pred_svc))
print('Training Accuracy on the test dataset:', accuracy_score(ytest, y_test_pred_svc))

# Display the confusion matrix and classification report for the test dataset
print('\nConfussion Matrix:',confusion_matrix(ytest, y_test_pred_svc))
print('\nClassification Report:',classification_report(ytest, y_test_pred_svc))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize the Gaussian Naive Bayes model
nb_model = GaussianNB()

# Train the model on training dataset
nb_model.fit(xtrain, ytrain)

# Make prediction on the train data set
y_train_pred_nb = nb_model.predict(xtrain)
# Make prediction on the test data set
y_test_pred_nb = nb_model.predict(xtest)

# Evaluate the model
print('Training Accuracy on the train dataset:', accuracy_score(ytrain, y_train_pred_nb))
print('Training Accuracy on the test dataset:', accuracy_score(ytest, y_test_pred_nb))

# Display the confusion matrix and classification report for the test dataset
print('\nConfussion Matrix:',confusion_matrix(ytest, y_test_pred_nb))
print('\nClassification Report:',classification_report(ytest, y_test_pred_nb))
```
